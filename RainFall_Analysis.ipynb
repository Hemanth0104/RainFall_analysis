ANNUAL RAINFALL RECEIVED BY EACH STATE OVER YEARS :

```python
plt.figure(figsize=(15,10))
sns.lineplot(x = 'YEAR' , y = 'ANNUAL', hue = 'STATE', data=df)
plt.title("ANNUAL RAINFALL RECEIVED", fontsize=20,)
```

I created a graph that depicted the yearly rainfall received by each state from 1901 to 2015. From this graph, we may get a basic idea of which states receive a lot of rain and which receive very little.

HIGHEST RAINFALL EVER RECEIVED IN A YEAR IN STATES :

```python
plt.figure(figsize=(15,15))
df.groupby(['STATE','YEAR'])['ANNUAL'].sum().sort_values(ascending=False).plot()
plt.grid()
plt.xlabel("State,Year", fontsize=15, color='blue')
plt.ylabel("Annual RAinFAll received",fontsize=15, color='orange')
plt.title("Highest RAinFAll in year data of states",fontsize=20, color="red")
```

The graphic shown displays an examination of the maximum amount of yearly rainfall received in the states from 1901 to 2015. From this, we may deduce that the largest quantity of rainfall ever received in the specified time period occurred in 1948, in the state of Arunachal Pradesh.

TOTAL AMOUNT OF RAINFALL RECIEVED OVERALL BY EACH STATE :

```python
plt.figure(figsize=(15,15))
df.groupby(['STATE'])['ANNUAL'].sum().sort_values(ascending=False).head(15).plot(kind='bar', color = 'blue')
plt.ylabel('Total Rainfall')
plt.title('Total Rainfall Data',fontsize=20,color='r')
plt.grid()
```

The graphic in this case displays a study of the total quantity of yearly rainfall received in the states from 1901 to 2015. From this, we may conclude that Coastal Karnataka receives the most rainfall overall.

THE MONTH WITH HIGHEST RAINFALL :
    

```python
plt.figure(figsize=(10,7))
df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG',
       'SEP', 'OCT', 'NOV', 'DEC']].mean().plot(kind= 'bar', color='red')
plt.xlabel('Months',fontsize=15)
plt.ylabel('Avg. Rainfall',fontsize=15)
plt.title('Avg. Monthly Rainfall Data',fontsize=25)
plt.grid()
plt.show()
```

According to the graph, July gets the greatest rainfall on average. In comparison, December has the least amount of rain. Furthermore, rainfall amounts vary greatly between months.

SEASON WISE RAINFALL IN INDIA :

```python
plt.figure(figsize=(10,2))
df[['STATE', 'Jan-Feb', 'Mar-May',
       'Jun-Sep', 'Oct-Dec']].groupby("STATE").mean().sort_values('Jun-Sep').plot.bar(width=0.5,edgecolor='k',align='center',stacked=True,figsize=(16,8))
plt.xlabel('STATE',fontsize=15)
plt.ylabel('Rainfall (in mm)',fontsize=15)
plt.title('Rainfall in States of India',fontsize=25)
plt.grid()
```

The graph above displays the average rainfall received in each season in various Indian states. From this, we can deduce which state in India receives the most rainfall and observe which season our nation receives the most rainfall, which is June-September, and so we arranged the graph according to the average quantity of rainfall in this season.

HEAT MAP OF RAINFALL

```python
fig=plt.gcf()
fig.set_size_inches(15,10)
fig=sns.heatmap(df.corr(),annot=True,linewidths=1,linecolor='k',square=True,mask=False, vmin=-1, vmax=1,cbar_kws={"orientation": "vertical"},cbar=True)
```

From this, we can observe that annual rainfall has a very strong association with rainfall obtained between June and September.

ANNUAL RAINFALL FROM YEAR 1901 TO 2015 IN INDIA

```python
df.groupby("YEAR").mean()['ANNUAL'].plot(ylim=(1000,2000),color='k',marker='o',markerfacecolor='red',linestyle='-',linewidth=2,figsize=(12,10))
plt.xlabel('Year',fontsize=20)
plt.ylabel('Annual Rainfall (in mm)',fontsize=20)
plt.title('Annual Rainfall from Year 1901 to 2015 in India',fontsize=25)

plt.grid()
```

The graph above displays the average rainfall received in India each year between 1901 and 2015. As a result, we can summarise the years in India when there was very little rainfall and when there was a lot of rain.

 BOX PLOT OF ANNUAL RAINFALL IN STATES OF INDIA :

```python
plt.rcParams['figure.figsize']=(18,8)
ax = sns.boxplot(x="STATE", y="ANNUAL", data=df, width=0.8,linewidth=3)
ax.set_xlabel('State',fontsize=20)
ax.set_ylabel('Annual Rainfall (in mm)',fontsize=20)
plt.title('Annual Rainfall in States of India',fontsize=30)
ax.tick_params(axis='x',labelsize=10,rotation=90)
ax.tick_params(axis='y',labelsize=10,rotation=0)
plt.grid()
```

From this, we can see that Arunachal Pradesh has the greatest annual rainfall, but its median is close to Coastal Karnataka, which indicates their average total yearly rainfall across years is around the same, but Arunachal Pradesh has the highest annual rainfall.
